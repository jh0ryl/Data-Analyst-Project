```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_Phil = df[(df['job_country'] == 'Philippines') & (df['job_title_short'] == 'Data Analyst')].copy()
```

```python
df_DA_Phil['job_posted_month_no'] = df_DA_Phil['job_posted_date'].dt.month
```

```python
df_DA_Phil_explode = df_DA_Phil.explode('job_skills')
```

```python
df_DA_Phil_pivot = df_DA_Phil_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0) 

df_DA_Phil_pivot.loc['Total'] = df_DA_Phil_pivot.sum()

df_DA_Phil_pivot = df_DA_Phil_pivot[df_DA_Phil_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_Phil_pivot = df_DA_Phil_pivot.drop('Total')  

df_DA_Phil_pivot

```

```python
df_DA_Phil_total = df_DA_Phil.groupby('job_posted_month_no').size()
df_DA_Phil_perc = df_DA_Phil_pivot.div(df_DA_Phil_total/100, axis=0) 

df_DA_Phil_perc = df_DA_Phil_perc.reset_index()
df_DA_Phil_perc['job_posted_month'] = df_DA_Phil_perc['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_Phil_perc = df_DA_Phil_perc.set_index('job_posted_month')
df_DA_Phil_perc = df_DA_Phil_perc.drop(columns='job_posted_month_no')

df_DA_Phil_perc
```

```python
from matplotlib.ticker import PercentFormatter
import numpy as np

# take first 5 skills for plotting
df_plot = df_DA_Phil_perc.iloc[:, :5]

# set the theme and plot (capture the axes object)
sns.set_theme(style='ticks')
plt.figure(figsize=(9, 5))
ax = sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.despine()
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2025')
plt.tight_layout()

# format y-axis as percentages; data are in 0-100 range so set xmax=100
ax.yaxis.set_major_formatter(PercentFormatter(xmax=100))

# remove legend (if present)
if ax.get_legend() is not None:
    ax.get_legend().remove()

# place skill labels to the right of the last point with small vertical offsets
# compute a vertical offset based on axis range to avoid overlapping labels
y_min, y_max = ax.get_ylim()
y_range = y_max - y_min
offset_step = y_range * 0.03  # 3% of y-range per step
x_right = ax.get_xlim()[1]
for i, col in enumerate(df_plot.columns):
    y = df_plot.iloc[-1, i]
    # center offsets around zero so labels spread both up and down
    offset = (i - (len(df_plot.columns)-1)/2) * offset_step
    ax.text(x_right + 0.5, y + offset, col, va='center')

plt.show()
```
